### Visuals

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
ax.scatter(h_data['Id'],h_data['SalePrice'])
plt.title("Homes' Prices")
plt.xlabel("Homes' Id record")
plt.ylabel('Prices')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
ax.scatter(h_data['Id'],h_data['LotArea'])
plt.title('Lot size in square feet')
plt.xlabel("Homes' Id record")
plt.ylabel('Lot Area')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
ax.scatter(h_data['Id'],h_data['GrLivArea'])
plt.title("Ground living area square feet")
plt.xlabel("Homes' Id record")
plt.ylabel("Ground Living Area")
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
ax.hist(h_data['BedroomAbvGr'], bins=30)
plt.title('Number of bedrooms per homes')
plt.xlabel('Bedrooms')
plt.ylabel("Number of Homes")
plt.show()
```
